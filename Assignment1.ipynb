In this assignment students have to transform iris data into 3 dimensions  and plot a 3d chart with transformed dimensions and colour each data  point with specific class.

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition 
from sklearn import datasets
import pandas as pd
```

```python
X = datasets.load_iris().data
y = datasets.load_iris().target
```

```python
pca = decomposition.PCA(n_components=3)
transformed_data = pca.fit_transform(X)
```

```python
fig = plt.figure()
ax = Axes3D(fig, elev = -150, azim = 100)
ax.scatter(transformed_data[:,0],transformed_data[:,1],transformed_data[:,2], c=y, cmap = plt.cm.autumn_r)
ax.set_title('Iris Dataset')
ax.set_xlabel('1st Eigen Vector')
ax.set_ylabel('2nd Eigen Vector')
ax.set_zlabel('3rd Eigen Vector')
ax.legend(['setosa', 'versicolor', 'virginica'])
plt.show()
```
